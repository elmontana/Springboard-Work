```python
import pandas as pd
import numpy as np
```

```python
import json
from pandas.io.json import json_normalize
```

```python
data = json.load((open('data/world_bank_projects.json')))
```

```python
df = json_normalize(data, 'majorsector_percent', 'mjsector_namecode')
```

```python
df = json_normalize(data)
df
```

```python
#Solution to First Problem, "Find the 10 countries with most projects"
countries = df.groupby('countryshortname').size().sort_values(ascending=False)
countries.head(10)
```

```python
themes = json_normalize(data, 'mjtheme_namecode')
```

```python
t1 = themes.sort_values(['code', 'name'])
```

```python
t1 = t1.replace(to_replace = {'name':{'':np.nan}}).fillna(method='bfill')
```

```python
t1.groupby('name').size().sort_values(ascending=False).head(10)
```
